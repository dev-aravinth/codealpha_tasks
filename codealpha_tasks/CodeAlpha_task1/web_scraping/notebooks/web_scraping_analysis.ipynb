```python
import pandas as pd
import matplotlib.pyplot as plt

quotes = pd.read_csv('../datasets/quotes_dataset.csv')
books  = pd.read_csv('../datasets/books_dataset.csv')
print('Quotes shape:', quotes.shape)
print('Books shape:', books.shape)
```

## 1. Quotes Dataset â€” Basic EDA

```python
quotes.head()
```

```python
quotes['author'].value_counts().head(10).plot(kind='barh', title='Top 10 Most Quoted Authors')
plt.tight_layout()
plt.show()
```

## 2. Books Dataset â€” Basic EDA

```python
books.describe()
```

```python
books['price_gbp'].hist(bins=20, edgecolor='black')
plt.title('Price Distribution')
plt.show()
```

```python
books['rating_stars'].value_counts().sort_index().plot(kind='bar', rot=0)
plt.title('Books by Rating')
plt.show()
```
